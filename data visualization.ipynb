```python
import pandas as pd
df=pd.read_csv('/home/msois/Downloads/Accidents(1)/Accidents/Accident_Cities.csv')
df.head()
```

```python
#1 max number of people injures 11-15
df['injured']=df['Persons Injured - 2011']+df['Persons Injured - 2012']+df['Injured - 2014']+df['Injured - 2015']
df1=df.sort_values("injured", ascending = False) 
df1.head(3)
```

```python
#2 injury decreased in 2011
df[df["Persons Injured - 2011"].ge(df["Persons Injured - 2011"])].head()
```

```python
#4 average death12-16
df["death rate"] =(df["Persons Killed - 2011"] + df["Persons Killed - 2012"] + df["Killed - 2013"] + df["Killed - 2014"] + df["Killed - 2015"])
df
```

```python
#5
df2=df.sort_values("death rate", ascending = False) 
df2.head(5)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.bar(df2['Name of City'], df2['death rate'])
plt.xticks(df2['Name of City'])
plt.xlabel('City')
plt.ylabel('Death')
plt.title('Max death')

plt.show()
```
